```python
import pandas as pd
```

```python
# read the csv file into a pandas dataframe
df_tuitionAdd = pd.read_csv('tuition and additional compulsory fees.csv')
df_tuitionAdd
```

```python
# getting information about dataset
df_tuitionAdd.info()
```

```python
# For better understanding information in the dataset to print the unique values and their counts for the column
# loop through each column in the dataframe
for column in df_tuitionAdd.columns:
    print(df_tuitionAdd[column].value_counts())
```

```python
# drop unrelated/blank columns
df_tuitionAdd.drop(['DGUID','UOM','UOM_ID','SCALAR_FACTOR','SCALAR_ID','VECTOR','COORDINATE','STATUS','SYMBOL','TERMINATED','DECIMALS'], axis=1, inplace=True)
df_tuitionAdd
```

```python
# drop null values
df_tuitionAdd.dropna(inplace=True)
df_tuitionAdd
```

```python
# save the dataframe as csv file
df_tuitionAdd.to_csv('cleaned_tuition_additional_compulsory_fees.csv', index=False)
```
