```python
# read the csv file into a pandas dataframe
df_medianIncome = pd.read_csv('cleaned_median_employment_income_years.csv')
```

```python
# generating a Seaborn barchart to analyze postgraduate income by years
sns.barplot(y=df_medianIncome['Field of study'],x=df_medianIncome['VALUE'],hue=df_medianIncome['Graduate statistics'],palette="magma",errorbar=('ci', 20))
plt.legend(bbox_to_anchor=(1, 1),loc='upper left',borderaxespad=0)
plt.title('Median employment income after two, five, ten years of graduation')
plt.xlabel('Income')
```

```python
# generating a Seaborn barchart to analyze postgraduate income by degree
sns.barplot(y=df_medianIncome['Field of study'],x=df_medianIncome['VALUE'],hue=df_medianIncome['Educational qualification'],palette="magma_r",errorbar=('ci', 0))
plt.legend(bbox_to_anchor=(1, 1),loc='upper left',borderaxespad=0)
plt.title('Median employment income depending on educational qualification')
```

```python
# generating a Seaborn barchart to compare fees for undergraduate and graduate education
sns.barplot(x=df_tuitionAdd['REF_DATE'],y=df_tuitionAdd['VALUE'],hue=df_tuitionAdd['Level of study'],palette="rocket",errorbar=('ci', 20))
plt.legend(bbox_to_anchor=(1, 1),loc='upper left',borderaxespad=0)
plt.title('Tuition and additional compulsory fees')
plt.xlabel('Years')
plt.ylabel('Fees')
```

```python
# generating a Seaborn barchart to fees by field of study
sns.barplot(y=df_tuitionUgr['Field of study'],x=df_tuitionUgr['VALUE'], palette="twilight",errorbar=('ci', 0))
plt.title('Average tuition fees for undergraduate students by field of study')
plt.xlabel('Fees')
```

```python
# generating a Seaborn barchart to fees by provinces
sns.barplot(y=df_tuitionUgr['GEO'],x=df_tuitionUgr['VALUE'], palette="twilight",errorbar=('ci', 0))
plt.title('Average tuition fees for undergraduate students by geography')
plt.xlabel('Fees')
plt.ylabel('Provinces')
```
